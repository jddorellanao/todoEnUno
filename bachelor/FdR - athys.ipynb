Athy's Law

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science', 'notebook', 'grid'])
```

```python
x=[0.38,0.35,0.33,0.31,0.27,0.26,0.24,0.23,0.21,0.2] #porosidad
y=[300,600,900,1200,1500,1800,2100,2400,2700,3000] #profundidad
```

```python
lphi=np.log(x) #log(phi)
zlphi=np.multiply(y,lphi) #(z)log(phi)
z2=np.square(y) #z^2
```

```python
n=len(x)
sz=np.sum(y)
slphi=np.sum(lphi)
szlphi=np.sum(zlphi)
sz2=np.sum(z2)
```

```python
m=(n*szlphi-sz*slphi)/(n*sz2-(sz)**2)
b=(sz2*slphi-sz*szlphi)/(n*sz2-(sz)**2)
```

```python
k=-m
phi0=np.exp(b)

print('el valor de k es igual a '+str(k))
print('el valor de phi 0 es igual a '+str(phi0))
```

```python
phi=np.zeros([n])

for a in range(0,n):
    phi[a]=phi0*np.exp(-k*y[a])

fig, ax = plt.subplots()
ax.plot(phi,y,'-r')
ax.plot(x,y,'Hb')
ax.invert_yaxis()

plt.xlabel('Porosidad $\it{\phi}$')
plt.ylabel("Profundidad $\it{[m]}$")
plt.title('Porosidad vs Profundidad', fontsize=20)
plt.grid(True)
```
