Modelo Budiansky

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science', 'notebook', 'grid'])
```

```python
ks = 3.7e10
ki = 0.22e10
mus = 4.4e10
mui = 0
rhom = 2.65e3 
rhof = 1e3 
alfa = 3e-1
ns=0.5*(3*ks-2*mus)/(3*ks+mus)
phi=np.zeros([15])
Vp=np.zeros([15])
Vs=np.zeros([15])
```

```python
for b in range(0,15):
    phi[b]=b/100
    epsilon=(b/100*3)/(4*np.pi*alfa)
    nsc=ns*(1-16*epsilon/9)
    kb=ks*(1-(16*(1-nsc**2)*epsilon)/(9*(1-2*nsc)))
    mub=mus*(1-(32*(1-nsc)*(5-nsc)*epsilon)/(45*(2-nsc)))
    rhobulk=(1-b/100)*rhom+b/100*rhof
    Vp[b]=((3*kb+4*mub)/(3*rhobulk))**0.5
    Vs[b]=(mub/rhobulk)**0.5
```

```python
fig, ax = plt.subplots()
ax.plot(phi,Vp,label='Vp')
ax.plot(phi,Vs,label='Vs')
legend = ax.legend(loc='best', shadow=True, fontsize='x-large')

plt.xlabel("Porosidad $\it{\phi}$")
plt.ylabel("Velocidad (m/s)")
plt.title("Modelo Budiansky and O'Connell", fontsize=20)
plt.grid(True)
```
