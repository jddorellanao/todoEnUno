Porosidad-Permeabilidad

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use(['science', 'notebook', 'grid'])
```

```python
x=np.array([0.020,0.020,0.111,0.096,0.156,0.150,0.075,0.105,0.060,0.179,0.156]) #porosidad
y=np.array([0.01,0.02,22,10.5,135.6,120,11,15.6,0.6,350,130]) #kh
xy=np.multiply(x,y) #x*y
x2=np.square(x) #x^2
z=np.linspace(620,642,11) #profundidad
N=len(x)
```

```python
#Calculos
X=np.log(x) #log(x)
Y=np.log(y) #log(y)
XY=np.log(xy) #log(x*y)
X2=np.log(x2) #log(x^2)
xl2=np.square(X) #(log(x))^2

slx=np.sum(X) #sum logx SE UTILIZA
sly=np.sum(Y) #sum logy SE UTILIZA
slxly=np.sum(X*Y) #sum logx*logy SE UTILIZA
sx2=np.sum(x2) #sum logx^2
sxy=np.sum(xy) #sum x*y
slxs=np.sum(X2) #sum logx^2
slxy=np.sum(XY) #sum logxy
slx2=np.sum(xl2) #sum (logx)^2 SE UTILIZA

n=(N*slxly-slx*sly)/(N*slx2-slx**2)
b=(slx2*sly-slxly*slx)/(N*slx2-slx**2)
M=np.exp(b)
```

```python
#b=13.2442
#n=4.5023
#M=np.exp(b)
phi=np.zeros([11])
k=np.zeros([11])
c=0
for a in range(0,22,2):
    phi[c]=a/100
    k[c]=M*(phi[c]**n)
    c+=1

#plt.figure(1)
fig, ax = plt.subplots()
ax.loglog(phi,k,'-r')
ax.loglog(x,y,'Hb')

#plt.figure(2)
#fig, ax = plt.subplots()
#ax.plot(phi,k,'-r')
#ax.plot(x,y,'Hb')

plt.xlabel('Porosidad $\it{\phi}$')
plt.ylabel("Permeabilidad $\it{[mD]}$")
plt.title('Porosidad vs Permeabilidad', fontsize=20)
plt.grid(True)
```
